```python
import seaborn as sns

tips = sns.load_dataset('tips')
iris = sns.load_dataset('iris')
titanic = sns.load_dataset('titanic')
planets = sns.load_dataset('planets')
```

```python
sns.scatterplot(x="tip", y="total_bill", data=tips, hue="day", size="size", palette="YlGnBu")
```

```python
sns.histplot(tips["tip"], kde=True, bins=15)
#kde makes the line appear
```

```python
sns.displot(tips['tip'], kde=True, bins=15)
```

```python
sns.barplot(x="sex",y="tip",data=tips)
```

```python
sns.boxplot(x="day", y="tip", data=tips, palette="YlGnBu")
```

```python
sns.stripplot(x="day", y="tip", data=tips, hue="sex", dodge=True)
```
